### Data

**Immigration to Canada data**

```python
cananda_file = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx'

df_can_imigration = (pd.read_excel(cananda_file
                                  , sheet_name='Canada by Citizenship'
                                  , skiprows=range(20)
                                  , skipfooter=2
                                 )
                     .rename({'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, axis=1)
                     .set_index('Country')
                     #.assign
                    )

print (f'Imigration count: {len(df_can_imigration):,}')
df_can_imigration
```

### Line Plot

```python
## Make column names as string
df_can_imigration.columns = list(map(str, df_can_imigration.columns))
years = list(map(str, range(1980, 2014)))

years
```

#### Imigration by country

```python
yr = df_can_imigration.filter(years, axis=1)

### transpose
fig = px.line(yr.transpose())
fig.show()
```

#### Country specific (Haiti) 

```python
haiti = (df_can_imigration.query('Country == "Haiti"')
         .filter(years, axis=1)
        )

haiti
```

```python
fig = px.line(haiti.transpose())
fig.show()
```

```python
df_china_india = (df_can_imigration
                  .query('Country == "China" | Country == "India"')
                  .filter(years, axis=1)
                  .transpose()
                 )

df_china_india
```

```python
fig = px.line(df_china_india)
fig.show()
```

#### Top5 Imigration

```python
#years.append('Total')
df_top5_imigraiton = (df_can_imigration
                     .assign(Total=lambda df: df.sum(axis=1))
                     .filter(years, axis=1)
                     .sort_values('Total', ascending=False)
                     .head(5)
                     .drop('Total', axis=1)
                     .transpose()
                     )


df_top5_imigraiton.head(2)
```

```python
fig = px.line(df_top5_imigraiton
              , title="<b>Top 5 countries Imigration Trends</b>"
              , labels={'value': '<b>values</b>', 'index': '<b>Year</b>'}
              , template='plotly_white'
             )
fig.show()
```

```python
df_top5_imigraiton.plot(kind="area", stacked=False)
```

## Scatter Plot

### Iris data

```python
iris_df = px.data.iris()
iris_df.head(2)
```

```python
fig = px.scatter(iris_df, x="sepal_width", y="petal_width"
                 , color="species"
                 , size="petal_length"
                )
fig.show()
```

```python
fig_iris = go.Figure()

fig_iris.add_trace(go.Scatter(x=iris_df.sepal_width, y=iris_df.petal_width
                             , mode='lines+markers'
                             , name='1'
                             , marker =dict(size=iris_df.petal_length)
                        
                        )
             )

fig_iris.add_trace(go.Scatter(x=iris_df.sepal_width, y=iris_df.sepal_length
                              , mode='markers'
                              , name='2'
                              , marker =dict(color=iris_df.petal_length)
                        )
             )


fig_iris.show()
```

```python
import numpy as np
np.random.seed(1)

N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5

fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=random_x, y=random_y0,
                    mode='markers',
                    name='markers'))
fig.add_trace(go.Scatter(x=random_x, y=random_y1,
                    mode='lines+markers',
                    name='lines+markers'))
fig.add_trace(go.Scatter(x=random_x, y=random_y2,
                    mode='lines',
                    name='lines'))

fig.show()
```
